```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
fig = plt.figure()
plt.plot(x)
plt.title('test')     # 設定 title
plt.show()

x = [1,2,3,4,5]
fig, ax = plt.subplots()
ax.set_title('test')  # 設定 title
ax.plot(x)
plt.show()

```

```python
import numpy as np
from matplotlib import pyplot as plt

x = np.arange(1,11)
y = 2* x +5
plt.title("demo")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,y)
plt.show
```
